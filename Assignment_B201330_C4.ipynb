# Preprocessing

```python
# Preprocessing
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

df = pd.read_csv("train_titanic.csv")
df_model = df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])

imputer = SimpleImputer(strategy='most_frequent')
df_model[['Age', 'Embarked']] = imputer.fit_transform(df_model[['Age', 'Embarked']])

label_encoders = {}
for col in ['Sex', 'Embarked']:
    le = LabelEncoder()
    df_model[col] = le.fit_transform(df_model[col])
    label_encoders[col] = le

X = df_model.drop('Survived', axis=1)
y = df_model['Survived']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# Simple Linear Regression

```python
# Simple Linear Regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_fare = X_train[:, [5]]  # Using only 'Fare' feature
X_fare_test = X_test[:, [5]]

model_slr = LinearRegression()
model_slr.fit(X_fare, y_train)
y_pred_slr = model_slr.predict(X_fare_test)

mse_slr = mean_squared_error(y_test, y_pred_slr)
r2_slr = r2_score(y_test, y_pred_slr)

print("Simple Linear Regression MSE:", mse_slr)
print("Simple Linear Regression R² Score:", r2_slr)
```

```python
# Plotting Simple Linear Regression
plt.figure(figsize=(8, 5))
plt.scatter(X_fare_test, y_test, color='blue', label='Actual')
plt.plot(X_fare_test, y_pred_slr, color='red', linewidth=2, label='Regression Line')
plt.title('Simple Linear Regression: Fare vs Survived')
plt.xlabel('Standardized Fare')
plt.ylabel('Survived')
plt.legend()
plt.grid(True)
plt.show()
```

# Multiple Linear Regression

```python
# Multiple Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model_mlr = LinearRegression()
model_mlr.fit(X_train, y_train)
y_pred_mlr = model_mlr.predict(X_test)

mse_mlr = mean_squared_error(y_test, y_pred_mlr)
r2_mlr = r2_score(y_test, y_pred_mlr)

print("Multiple Linear Regression MSE:", mse_mlr)
print("Multiple Linear Regression R² Score:", r2_mlr)
```

```python
# Plotting Residuals for Multiple Linear Regression
import matplotlib.pyplot as plt

residuals = y_test - y_pred_mlr

plt.figure(figsize=(8, 5))
plt.scatter(y_pred_mlr, residuals, color='purple', alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--')
plt.title("Residuals Plot: Multiple Linear Regression")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.grid(True)
plt.show()
```

# Decision Tree Classifier

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

model_dt = DecisionTreeClassifier(random_state=42)
model_dt.fit(X_train, y_train)
y_pred_dt = model_dt.predict(X_test)

acc_dt = accuracy_score(y_test, y_pred_dt)
cm_dt = confusion_matrix(y_test, y_pred_dt)
cr_dt = classification_report(y_test, y_pred_dt)

print("Decision Tree Accuracy:", acc_dt)
print("Confusion Matrix:\n", cm_dt)
print("Classification Report:\n", cr_dt)
```

# K-Means

```python
#K-Means
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
df = pd.read_csv("train_titanic.csv")
df = df.dropna()
X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']].copy()  # Make a copy
y = df['Survived']

label_encoder = LabelEncoder()
X.loc[:, 'Sex'] = label_encoder.fit_transform(X['Sex'])  # Use .loc to avoid the warning

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)
cluster_labels = kmeans.predict(X_scaled)
print(f"Inertia: {kmeans.inertia_}")

y_true = df['Survived']
cm = confusion_matrix(y_true, cluster_labels)
print("Confusion Matrix:")
print(cm)

cr = classification_report(y_true, cluster_labels)
print("Classification Report:")
print(cr)

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[cluster_labels == 0, 0], X_pca[cluster_labels == 0, 1], s=100, c='blue', label='Cluster 1')
plt.scatter(X_pca[cluster_labels == 1, 0], X_pca[cluster_labels == 1, 1], s=100, c='red', label='Cluster 2')

centroids = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='yellow', marker='x', label='Centroids')

plt.title('Clusters of Passengers')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()


```
